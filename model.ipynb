```
import tensorflow as tf
import numpy as np
import pandas as pd
import csv
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
f = open("/content/Main_data.txt","r")
lines = f.readlines()
data = {}
cnt = 0
max1=0
keyDict = ""
dataArr = []
vocab = {}
# in
for line in lines:
    cnt +=1
    if cnt%3==1:
        # print(line)
        data[line.strip()[1:]] = {}
        keyDict = line.strip()[1:]
        dataArr.append([])
    elif cnt%3==2:
        data[keyDict]["sequence"] = line.strip()
        max1=max(max1,len(data[keyDict]["sequence"]))
        dataArr[-1].append(' '.join(list(data[keyDict]["sequence"])))
        # for i in range(len(dataArr[-1][-1])):
          # dataArr[-1][-1][i] = ord(dataArr[-1][-1][i]) - ord('A')
        for alphabet in (data[keyDict]["sequence"]):
            vocab[alphabet] = ord(alphabet) - ord("A")
    else:
        data[keyDict]["sequenceAns"] = line.strip()
        dataArr[-1].append(' '.join(list(data[keyDict]["sequenceAns"])))

    # print(line)
f.close()
print("max length is ",max1)
print(dataArr)
import csv
f = open("/content/data.csv","w")
writer = csv.writer(f)
writer.writerows(dataArr)
f.close()
```
